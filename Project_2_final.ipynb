## Scientific question
What is the molecular basis of naturally developed human CCR5 mutation in HIV resistance?

## C-C motif chemokine receptor type 5 (CCR5)
CCR5 is a member of the beta chemokine receptor family composed of 7 transmembrane region (similar structure to G-protein coupled receptors). It is expressed predominantely by T cells, macrophages, and dendritic cells and can bind and transduce CCL3, CCL4, and CCL5 signaling.

CCR5 also serves as an important coreceptor for human immunodeficiency virus (HIV) entry into target cells. The heterotrimeric complex formed by gp120 (expressed by HIV) binding to CD4 and CCR5 (expressed on target cells) facilitates fusion of viral capsid with target cell membrane. Previous research has shown that a mutation of 32-base-pair deletion (delta-32) in the coding region of CCR5 gene which introduced a premature stop codon resulted in reduced CCR5 expression and induced HIV resistance in people carrying this mutation (Samson et al., 1996b). Mutations in other regions of CCR5 gene have been discovered, and some of which were capable of affecting CCR5 expression and structure integrity and therefore bear the potential of inducing HIV resistance in individuals carrying the mutation (Carrington et al., 1997b)(Blanpain et al., 1999b).

## Hypothesis
If a mutation occurs on CCR5 that can influence the conformational integrity or structure of this molecule, which results in impaired CCR5 expression at cell surface or binding to gp120 expressed by HIV, then the cell carrying this mutation will be resistant to HIV infection.

### Methods
Structural bioinformatics was used to obtain the 3D structure of CD4-gp120-CCR5 complex to support the statement that CCR5 is an essential molecule involved facinilating HIV entrance into target cells. The structural data of CD4-gp120-CCR5 complex was obtained from PDB, and the the structure was also used to present 3D structure of wild-type CCR5 that was not bound to any inhibitors or ligands. 

Homology modeling was used to obtain the 3D structures of mutated CCR5 (delta-32, C20S, and C101A), of which the structural data could not be found online with designated PDB ID. Homology modeling was performed using SWISS-MODEL (https://swissmodel.expasy.org/) with protein sequences of the mutated CCR5. This allowed visualization of overall structural difference between wild-type and mutant CCR5.

Multiple sequence alignment was used to align sequence of multiple beta chemokine receptors with CCR5. This helped to determine whether the mutation in CCR5 examined occured in conserved region of the beta chemokine receptor family. Protein sequences used for multiple sequence alignment were found on NCBI. Homology modeling was performed using MAFFT online server (https://mafft.cbrc.jp/alignment/server/).

3D protein measurement was used to quantify difference in structures between the wild-type CCR5 and the mutated versions. Protein sequences were directly extracted from PDB files obtained in homology modeling.

Sequence logos was used to visualize the level of conservation at each amino acid position to determine if the mutations examined occured at conserved region of beta chemokine receptor family. The aligned protein sequences used for sequence logos were obtained from multiple sequence alignment.

```python
# Homology modeling performed by SWISS-MODEL to obtain protein structures of CCR5 mutant alleles. (https://swissmodel.expasy.org/)
# data download as pdb files
# define function show_pdb_file to create a more convenient way of visualizing all structures
# parameter file is the name of pdb file that need to be visualized
# parameter struc_format is the format of how the molecule will be presented, default presentation as cartoon
# parameter structure_sele is the part of the molecule to be presented, default presentation as protein
# parameter structure_color is the color the molecule will be presented in, default set to blue
# all parameters are local variables
def show_pdb_file(file, struc_format="cartoon", structure_sele="protein", structure_color="blue"):
    # visualize the struture in file and assign the visualized structure to local variable view
    view = nv.show_structure_file(file)
    # clear all presentation format of the structure
    view.clear()
    # present the structure in assigned format, selected portion, and color
    view.add_representation(struc_format, selection=structure_sele,color=structure_color)
    # show structure
    return view
```

## 3D protein measurement
3D protein measurement demonstrates features of proteins by quantfying the features through calculations performed based on the protein sequence. Comparing the result of 3D protein measurement allows a direct distinguish of structural difference between the proteins.

```python
# source code: https://towardsdatascience.com/visualizing-and-analyzing-proteins-in-python-bd99521ccd
# define a function protein_measurement to perform 3D protein measurement of proteins in pdb files
# parameter protein is the name of the protein to be analyzed
# parameter pdb_file is file to be read in and analyzed
# both parameters are local variables
def protein_measurement(protein, pdb_file):
    # extract protein sequence from the pdb file
    for record in SeqIO.parse(pdb_file, "pdb-atom"):
        # assign sequence to local variable sequence
        sequence=record.seq
    # uncomment below for code check
    # print(sequence)
    
    # convert sequence to the correct type for analysis and assign to local variable analyzed_seq
    analyzed_seq = ProteinAnalysis(str(sequence))
    # uncomment below for code check
    # print(type(analyzed_seq))
    
    # create an emtpy dictionary to store protein analysis result and assign the dictionary to local variable seq_analysis
    seq_analysis = {}
    # perform analysis
    # assign the name of analyzed protein as value to key sequence name
    seq_analysis['Sequence name'] = protein
    # assign the sequence of analyzed protein as value to key sequence
    seq_analysis['Sequence'] = str(sequence)
    # assign the length of sequence of analyzed protein as value to key sequence length
    seq_analysis['Sequence length'] = len(sequence)
    # assign the molecular weight of analyzed protein as value to key molecular weight
    seq_analysis['Molecular weight'] = analyzed_seq.molecular_weight()
    # assign the hydrophobicity level of analyzed protei as value to key GRAVY
    seq_analysis['GRAVY'] = analyzed_seq.gravy()
    # assign the secondary structure of analyzed protein (a tuple of helix, turn, and sheet) as value to key secondary structure
    seq_analysis['Secondary structure'] = analyzed_seq.secondary_structure_fraction()
    # uncomment below for code check
    # print(seq_info)
    
    # convert the dictionary into a pandas dataframe and assign dataframe to local variable seq_df
    seq_df=pd.DataFrame.from_dict(seq_analysis, orient='index')
    return seq_df
```

```python
# analyze wild-type CCR5 protein using defined function and assign result to global variable CCR5_analysis
CCR5_analysis=protein_measurement("CCR5", "CCR5.pdb")
# print result of analysis
print(CCR5_analysis)
```

```python
# analyze CCR5 protein with delta-32 mutation using defined function and assign result to global variable delta_32_analysis
delta_32_analysis=protein_measurement("delta_32", "delta_32.pdb")
# print result of analysis
print(delta_32_analysis)
```

```python
# analyze CCR5 with C20S mutation using defined function and assign result to global variable C20S_analysis
C20S_analysis=protein_measurement("C20S", "C20S.pdb")
# print result of analysis
print(C20S_analysis)
```

```python
# analyze CCR5 with C101A mutation using defined function and assign result to global variable C101A_analysis
C101A_analysis=protein_measurement("C101A", "C101A.pdb")
# print result of analysis
print(C101A_analysis)
```

## Sequence logos
Sequence logos graphically demonstrates conserved regions between different strands of proteins or nucleotides.

```python
# create an empty list and assign the list to global variable seq_list
seq_list = list()
# append all alignment sequences from the alignment data to seq_list
for aln in alignments:
    sequence=str(aln.seq)
    seq_list.append(sequence)
# uncomment below for code check
# print(seq_list)
```

```python
# source code: https://logomaker.readthedocs.io/en/latest/examples.html
# create a matrix containing frequency of each amino acid at each position using the list created previously
# assign the matrix to global variable aln
aln=lg.alignment_to_matrix(sequences=seq_list, to_type='counts', characters_to_ignore='.-X')
# uncomment below for code check
# print(aln)
```

```python
# plot sequence logos graph using the matrix created previously
lg.Logo(aln)
```

## Result analysis
Structure of CCR5 with C20S or C101A mutation does not show great variation from the wildtype. Structure of CCR5 with delta-32 mutation shows that this mutation will cause this protein to lose its conformational integrety. Since CCR5 is required for HIV entry into the cell by forming CD4-gp120-CCR5 complex, delta-32 mutation is expected to hinder the function of CCR5 as coreceptor for HIV entrance. This could result in the development of HIV resistance, which is proven by research demonstrating that people with CCR5-delta-32 allele are resitant to HIV infection. Although C20S and C101A mutation do not seeme to influence the overall secondary structure of CCR5 based on result from homology modeling, point mutation in cystine could result in a loss of disulfide bond, which could also influence either the structure or the conformational integrity of the protein. Therefore, these two mutation could potentialy hinder the ability of CCR5 serving as coreceptor of HIV entry into host cells. This is also proven by research demonstrating that cells carrying genes with C20S mutation or C101A mutation have reduced expression level of CCR5 at cell surface, which in turn reduced HIV infection. Therefore, it is likely that a mutation in CCR5 which could alter the structure of the protein and reduce its expression at cells surface could cause the development of HIV resistance. Interestingly, the two point mutations analyzed, C20S and C101A, both occured in highly conserved region throughout beta-chemokine receptor. This could indicate that the mutations could potentialy have larger impact on CCR5 function than expected from the strucutural difference demonstrated by homology modeling and 3D protein measurement.
